```python
import numpy as np
import pandas as pd
```

```python
ssps = ['SSP126', 'SSP245', 'SSP370', 'SSP585']

fairgp_scores = pd.read_csv('../output/ssps/FaIRGP/cv-scores.csv')
fairgp_scores.index = ssps
fairgp_scores.index.name = 'SSP'

gp_scores = pd.read_csv('../output/ssps/plain-GP/cv-scores.csv')
gp_scores.index = ssps
gp_scores.index.name = 'SSP'

fair_scores = pd.read_csv('../output/ssps/FaIR/cv-scores.csv')
fair_scores.index = ssps
fair_scores.index.name = 'SSP'
```

```python
fairgp_summary = fairgp_scores.aggregate(['mean', 'std'])
gp_summary = gp_scores.aggregate(['mean', 'std'])
fair_summary = fair_scores.aggregate(['mean', 'std'])

fairgp_scores = pd.concat([fairgp_scores, fairgp_summary])
gp_scores = pd.concat([gp_scores, gp_summary])
fair_scores = pd.concat([fair_scores, fair_summary])
```

```python
fairgp_scores['Model'] = '2FaIRGP'
gp_scores['Model'] = '1Plain GP'
fair_scores['Model'] = '0FaIR'
metrics_cols = ['Bias', 'RMSE', 'MAE', 'Corr', 'LL', 'Calib95', 'CRPS', 'ICI']
full_scores = pd.concat([fairgp_scores, gp_scores, fair_scores]).sort_index().set_index('Model', append=True)
full_scores.columns = metrics_cols
full_scores = full_scores[['RMSE', 'MAE', 'Bias', 'Corr', 'LL', 'Calib95', 'CRPS']].sort_index().round(2)
```

```python
results = full_scores[['RMSE', 'MAE', 'Bias', 'Corr', 'LL', 'Calib95', 'CRPS']].fillna('-')
# results.to_latex('./tables/ssp-experiment-global-scores.tex')
results
```
